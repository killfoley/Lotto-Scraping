### 2021

```python
#import the library used to query a website
import requests
#specify the url for 2021
lottery = "https://www.irishlottery.com/euromillions-archive-2021"
#Query the website and return the html to the variable 'page'
page = requests.get(lottery)
page
```

```python
page.status_code
```

```python
#import the Beautiful soup functions to parse the data returned from the website
from bs4 import BeautifulSoup
#Parse the html in the 'page' variable, and store it in Beautiful Soup format
soup = BeautifulSoup(page.content, 'lxml')
```

```python
# Find the 5 lotto numbers available at the class = "Ball"
ball_class_21 = soup.find_all("li", class_="ball")
#get the text using list comprehension
ball_21 = [pt.get_text() for pt in ball_class_21]
type(ball_21)
```

```python
#convert into a dataframe in acsending order
import numpy as np
import pandas as pd
df_ball_21 = pd.DataFrame(np.array(ball_21).reshape(-1,5),columns = list("ABCDE"))
df_ball_21
```

```python
# Find the lucky star numbers available at the class = "lucky-star"
ball_lucky_star_class_21 = soup.find_all("li", class_="lucky-star")
ball_lucky_star_21 = [pt.get_text() for pt in ball_lucky_star_class_21]
type(ball_lucky_star_21)
```

```python
df_lucky_stars_21 = pd.DataFrame(np.array(ball_lucky_star_21).reshape(-1,2),columns=['LStar1','LStar2'])
df_lucky_stars_21
```

```python
#join the two dataframes
df_2021 = pd.concat([df_ball_21, df_lucky_stars_21], axis=1)
df_2021.astype('int')
```

```python
df_ball = df_euromill.drop(['Year','LStar1','LStar2'], axis=1)
df_ball
```

```python
list_ball = list(df_ball.values.T.flatten())
```

```python
## The maximum occurance of number in the list of numbers
from collections import Counter
most_common,num_most_common = Counter(list_ball).most_common(1)[0]
```

```python
print(most_common)
print(num_most_common)
```

```python
# find the 5 most common numbers in the list
most_common_5 = Counter(list_ball).most_common(5)
most_common_5
```

```python
number_counter = {}
for number in list_ball:
    if number in number_counter:
        number_counter[number] += 1
    else:
        number_counter[number] = 1
 
popular_numbers = sorted(number_counter, key = number_counter.get, reverse = True)
 
print(popular_numbers)
print(number_counter)
```

```python
#print the dictionary with sorted values
dictionary_numbers = Counter(list_ball)
dictionary_numbers_sorted_keys = sorted(dictionary_numbers, key=dictionary_numbers.get, reverse=True)
for r in dictionary_numbers_sorted_keys:
    print ("{}\t: \t {} times".format(r, dictionary_numbers[r]))
```
